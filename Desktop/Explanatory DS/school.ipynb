```python
# Cleaning, Do we have empties? Do we have unwanted columns, Do we have unwanted row?
data.isnull().sum()
# Missing Data... Less than 2%, its Fine
```

```python
#Gender is a categorical variable with 0 for Male and 1 for Female. By filling null values with 2.0, we are creating a new category to represent unknown or unspecified gender
data['Gender'].fillna(2.0, inplace = True)
#Rank represents the class rank (1=Freshman, 2=Sophomore, 3=Junior, 4=Senior). Filling null values with 0.0 can indicate that the rank is unknown or not applicable. 
data['Rank'].fillna(0.0, inplace = True)
#Major is a nominal variable representing the student's field of study. Using 'Unknown' for missing values keeps the data categorical and allows for the identification of students who did not specify their major.
data['Major'].fillna('Unknown', inplace = True)
#smoking status is a nominal variable with values 0 for Nonsmoker, 1 for Past smoker, and 2 for Current smoker. Filling null values with 3.0 creates a new category to represent unknown smoking status
data['Smoking'].fillna(3.0, inplace = True)
```

```python
# Replacing
#we are replacing numerical codes with their corresponding categorical labels for several columns
data['Gender'].replace({0:'Male',
                        1:'Female',
                        2:'Neutral'},
                         inplace = True)

data['LiveOnCampus'].replace({0:'Off Campus',
                              1:'On Campus'},
                              inplace = True)

data['HowCommute'].replace({1:'Walk',
                              2:'Bike',
                              3:'Car',
                              4:'Public Transit',
                              5:'Other'},
                              inplace = True)
data['Rank'].replace({
    1:'Freshman',
2:'Sophomore',
3:'Junior',
4:'Senior',

},inplace=True)

data['Athlete'].replace({
    0:'Athlete',
    1:'Non-Athlete',
    
    },inplace=True)
data['Smoking'].replace({
    0:'Nonsmoker',
    1:'Pastsmoker',
    2:' Current smoker',
    
    },inplace=True)

# TODO - Rank, Gender, other categorical
```
